```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('ML_carsLIMPIO.csv', index_col='car_ID', dtype={"price": float})
df.sort_index()
```

```python
mP=np.median(df['price'])
mP
```

```python
#Definimos una nueva columna 'gama' a la cual le aplicamos 
#la siguiente condicion: si el precio es mayor a la mediana 
#colocar 1(alta gama) sino 0 (baja gama)
condicion = lambda x: '1' if x > mP else '0'
df['gama'] = df['price'].apply(condicion)
#df_sorted = df.sort_values(by='price')
#df_sorted.iloc[103]

```

```python
df.columns
```

```python
X=df[['doornumber','horsepower','enginesize']]
y=df['price']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42) 
```

```python
linear_model = LinearRegression()
linear_model.fit(X, y)
print('Las pendientes (coeficientes) son :', linear_model.coef_)
print('La ordenada al origen es:', linear_model.intercept_)
```

```python
from sklearn import metrics

y_pred=linear_model.predict(X_test)
print ('MAE:', metrics.mean_absolute_error(y_test, y_pred))
print ('MSE:', metrics.mean_squared_error(y_test, y_pred))
print ('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print ('R2:', metrics.r2_score(y_test, y_pred))
```
